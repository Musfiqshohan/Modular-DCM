```python
import numpy as np
import matplotlib.pyplot as plt
import os

# def addlabels(x,y):
#     for i in range(len(x)):
#         plt.text(i, y[i]//2,y[i], ha = 'center',
#                  Bbox = dict(facecolor = 'white', alpha = .5))
# function to add value labels
def addlabels(x,y,d):

	for i in range(len(x)):
		plt.text(i+d, y[i]+0.02, f'{y[i]}%', ha = 'center', rotation=0)

		# ax.text(i+d,y[i]+1, y[i], color='red',
        # bbox=dict(facecolor='none', edgecolor='red'))




# addlabels(x, y)

# set width of bar
barWidth = 0.200
fig = plt.subplots(figsize =(12, 8))

# set height of bar
list1 = dom1
list2 = intv
list3 = intv_aug

# Set position of bar on X axis
br1 = np.arange(len(dom1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]


plt.rcParams['font.size'] = 15
# ax.legend(loc='best', fontsize=25)


# Make the plot
plt.bar(br1, list1,  width = barWidth, label ='Domain1 Classifier')

# addlabels(br1, star_list, d=-0.05)


plt.bar(br2, list2,  width = barWidth,
		 label ='Interventional Classifier')


plt.bar(br3, list3,  width = barWidth,
		label ='Augmented Classifier')
addlabels(br2, list3, d=0.4)




# Adding Xticks
plt.xlabel('Facial attributes in CelebA dataset', fontweight ='bold', fontsize = 20)
plt.ylabel('Prediction accuracy (%)', fontweight ='bold', fontsize = 20)
plt.xticks([r + barWidth for r in range(len(list3))],
		labels, rotation=30)


plt.legend()

os.makedirs('/local/scratch/a/rahman89/PycharmProjects/EGSDE/runs/results', exist_ok='True')
# fig.savefig('/local/scratch/a/rahman89/PycharmProjects/EGSDE/runs/results/comparison.png', dpi=fig.dpi)
plt.savefig('/local/scratch/a/rahman89/PycharmProjects/EGSDE/runs/results/comparison.pdf', bbox_inches='tight')


plt.show()
```
